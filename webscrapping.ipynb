```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Start the webdriver
driver = webdriver.Chrome()

# Navigate to the Amazon product page
driver.get("https://www.amazon.com/dp/B08L5M9L4L?th=1")
driver.implicitly_wait(5)

# Click on the "See all reviews" link to load more reviews
more_review = driver.find_element(By.XPATH, '//*[@id="reviews-medley-footer"]/div[2]/a')
more_review.click()

time.sleep(5)

# Find the parent element using XPath
parent_element = driver.find_element(By.XPATH, "//div[@id='cm_cr-review_list']")

# Find child elements by class name within the parent element
child_elements = parent_element.find_elements(By.CLASS_NAME, "a-section.review.aok-relative")

# Create an empty list to store the data
data_list = []

for element in child_elements:
    # Extract name
    name = element.find_element(By.CLASS_NAME,'a-profile-name').text
    # Extract star rating
    star = element.find_element(By.XPATH, './/i[contains(@class, "a-icon-star")]/span').get_attribute("innerHTML")

    # Extract title, datetime, and description
    title = element.find_element(By.CLASS_NAME,'a-text-bold').text
    date_time = element.find_element(By.CLASS_NAME,'review-date').text
    desc = element.find_element(By.CLASS_NAME,'review-text-content').text
    
    # Append data to the list
    data_list.append([name, star, title, date_time, desc])

# Click on the "See all reviews" link to load more reviews
more_review = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
more_review.click()
time.sleep(5)

# Find the parent element using XPath
parent_element = driver.find_element(By.XPATH, "//div[@id='cm_cr-review_list']")

# Find child elements by class name within the parent element
child_elements = parent_element.find_elements(By.CLASS_NAME, "a-section.review.aok-relative")

for element in child_elements:
    # Extract name
    name = element.find_element(By.CLASS_NAME,'a-profile-name').text
    # Extract star rating
    star = element.find_element(By.XPATH, './/i[contains(@class, "a-icon-star")]/span').get_attribute("innerHTML")

    # Extract title, datetime, and description
    title = element.find_element(By.CLASS_NAME,'a-text-bold').text
    date_time = element.find_element(By.CLASS_NAME,'review-date').text
    desc = element.find_element(By.CLASS_NAME,'review-text-content').text
    
    # Append data to the list
    data_list.append([name, star, title, date_time, desc])

# Click on the "See all reviews" link to load more reviews
more_review = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
more_review.click()
time.sleep(5)

# Find the parent element using XPath
parent_element = driver.find_element(By.XPATH, "//div[@id='cm_cr-review_list']")

# Find child elements by class name within the parent element
child_elements = parent_element.find_elements(By.CLASS_NAME, "a-section.review.aok-relative")

for element in child_elements:
    # Extract name
    name = element.find_element(By.CLASS_NAME,'a-profile-name').text
    # Extract star rating
    star = element.find_element(By.XPATH, './/i[contains(@class, "a-icon-star")]/span').get_attribute("innerHTML")

    # Extract title, datetime, and description
    title = element.find_element(By.CLASS_NAME,'a-text-bold').text
    date_time = element.find_element(By.CLASS_NAME,'review-date').text
    desc = element.find_element(By.CLASS_NAME,'review-text-content').text
    
    # Append data to the list
    data_list.append([name, star, title, date_time, desc])


# Create a DataFrame from the list
df = pd.DataFrame(data_list, columns=['Name', 'Star', 'Title', 'Date Time', 'Description'])

# Save the DataFrame to a CSV file
df.to_csv('amazon_reviews.csv', index=False)

# Print the DataFrame
print(df)

# Close the webdriver when you're done
driver.quit()
```
